### Subplot in matplotlib

```python
x = np.linspace(1,5,11)
y = x** 2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, color = "r")

plt.subplot(1,2,2)
plt.plot(y,x, color = "b")
```

### Object oriented programming in matplotlib

```python
#This will create an imaginary figure
fig = plt.figure()
```

```python
fig = plt.figure()
#The set of list values represented the left side, bottom part of the figure, the width and the height of the figure
axes = fig.add_axes([0.1,0.1,0.8,0.8])

#Then we will be making the plot will axes and not with plt
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3,0.3])

axes1.plot(x,y,"k--")
axes1.set_xlabel("x axis")
axes1.set_ylabel("y axis")
axes1.set_title("small axis")

axes2.plot(y, x, "b-.")
axes2.set_xlabel("x axis")
axes2.set_ylabel("y axis")
axes2.set_title("Title")
fig.savefig("oop.jpeg")
```

```python
fig, axes = plt.subplots()

axes.plot(x,y)
```

```python
#subplots holds an arguments to specify the number of rows and columns for the plotting
fig,axes = plt.subplots(nrows =1, ncols =2)

#this line of codes below shows that axes is an array
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
#axes can also be selected by using indexing method
fig, axes = plt.subplots(nrows = 1, ncols =2)

axes[0].plot(x,x**2)
axes[0].set_title("first_plot")
axes[1].plot(x,x**3)
axes[1].set_title("Second plot")
```

```python
#plt.figure also holds two parameters which are: figsize and dpi
fig = plt.figure(figsize = (8,4), dpi=100)
#you can play around with the figsize to have different sizes for the figures

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
```

#### 
NB:

subplots always takes fig, axes

while,

figure takes fig and axes separately.

```python
fig, axes = plt.subplots(nrows = 1, ncols =2, figsize = (6,3))


axes[0].plot(x,y, label = "first")

axes[1].plot(y,x, label = "second")

plt.legend()
```

```python
fig = plt.figure(figsize = (8,6))

ax =fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label = "first")

ax.plot(y,x/3.0, label = "second")

ax.legend(loc = "center")
```
